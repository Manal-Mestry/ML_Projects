```python
# Drop columns with more than 50% missing data

threshold = len(df) * 0.5
cleaned_data = df.dropna(axis=1, thresh=threshold)

```

```python
# Remove remaining rows with null values
cleaned_data = cleaned_data.dropna()

# Display information about the cleaned dataset
cleaned_data.info(), cleaned_data.head()
```

```python
cleaned_data.isnull().sum()
```

```python
cleaned_data.head()
```

```python

# Convert 'Date' to datetime for time-based analysis
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'])

# Line plot for temperature trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='MeanTemp', data=cleaned_data, label='Mean Temp', color='blue')
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Mean Temperature (°C)')
plt.legend()
plt.show()

```

```python
# Bar chart for monthly precipitation
monthly_precip = cleaned_data.groupby('MO')['Precip'].count()
plt.figure(figsize=(12, 6))
monthly_precip.plot(kind='bar', color='teal')
plt.title('Monthly Precipitation Count')
plt.xlabel('Month')
plt.ylabel('Precipitation Count')
plt.show()
```

```python

# Histogram for temperature distribution
plt.figure(figsize=(12, 6))
sns.histplot(cleaned_data['MeanTemp'], bins=30, kde=True, color='orange')
plt.title('Distribution of Mean Temperature')
plt.xlabel('Mean Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```
